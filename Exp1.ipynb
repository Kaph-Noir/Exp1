```python
def get_financial_data(*names: str, rolling_sum: bool=False, rolling_window: int=4):
    for name in names:
        data = pd.read_csv(financial_items_path / f"{name}.csv", index_col="Period")
        data = data.filter(regex='R$', axis='index')
        data.dropna(how='all', axis='index', inplace=True)
        if rolling_sum:
            data = data.rolling(rolling_window).sum()
        yield data
```

```python
total_assets, net_assets, gross_profit = list(get_financial_data('자산총계', '자본총계', 'Gross Profit'))
```
